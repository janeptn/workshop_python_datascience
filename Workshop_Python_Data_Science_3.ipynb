```python
# Importando as bibliotecas

from sklearn.model_selection import train_test_split #divisão do conjunto de dados em treinamento e teste
from sklearn.neighbors import KNeighborsClassifier #algoritmo de classificação KNN
from sklearn.metrics import accuracy_score #métrica de avaliação
from sklearn import datasets #conjunto de dados
import pandas as pd #biblioteca pandas

import matplotlib.pyplot as plt #plotar gráficos
```

Classificação do Conjunto de Dados Íris - https://archive.ics.uci.edu/dataset/53/iris

image.png

```python
# Carregando o Conjunto de Dados

iris = datasets.load_iris()
```

```python
# Imprimindo as primeiras 5 linhas
print(iris.data[:5])
```

```python
# Imprimindo as colunas do dataframe
print('Features (variáveis)', iris.feature_names)

# Imprimindo as classes  do dataframe
print('Classes', iris.target_names)
```

```python
# Dividindo os Dados em Treino e Teste

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25, random_state=0) #75% para treinamento e 25% para teste
```

image.png

```python
# Treinando o Modelo KNN

knn = KNeighborsClassifier(n_neighbors=5) #número de vizinhos
knn.fit(X_train, y_train)
```

```python
# Realizando a Predição

y_pred = knn.predict(X_test)
```

```python
# Avaliando o Desempenho

accuracy = accuracy_score(y_test, y_pred)
print("Acurácia:", accuracy)
```

```python
# Plotando as amostras classificadas

cores = ["r", "g", "b"] #definindo as cores para cada classe

# Plotando as amostras
for i in range(len(y_test)):
    plt.scatter(X_test[i, 2], X_test[i, 3], c=cores[y_test[i]], marker="x") #plotando somente as 2 primeiras colunas

# Adicionando legendas
plt.legend(iris.target_names)

# Mostrando o gráfico
plt.show()
```
